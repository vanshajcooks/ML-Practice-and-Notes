```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04'],
    'City A': [15, 20, 18, 22],
    'City B': [10, 12, 15, 17]
}
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])  # Convert 'Date' to a datetime object


avg_city_a = df['City A'].mean()
avg_city_b = df['City B'].mean()
print(f"Average Temperature in City A: {avg_city_a}")
print(f"Average Temperature in City B: {avg_city_b}")


max_temp_date = df.loc[df['City A'].idxmax(), 'Date']
print(f"Date with highest temperature in City A: {max_temp_date}")


plt.plot(df['Date'], df['City A'], label='City A')
plt.plot(df['Date'], df['City B'], label='City B')
plt.title('Temperature Trends')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.legend()
plt.grid(True)
plt.show()

```
